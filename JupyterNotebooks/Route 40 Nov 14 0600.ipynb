```python
#As dicts aren't ordered, and order of stops is important OrderedDict is necessary
data = OrderedDict()
temp1 = df2.head(1)
c = int(temp1['Timestamp'])

for i in stops:
    temp = df2[df2['StopID'] == i ].head(1)
    StopID = int(temp['StopID'])
    b = int(temp['Timestamp'])
    seconds = (b-c)/1000000
    data[StopID] = seconds
print ("StopID and time in seconds")
print (data)
    

#new_df.append(data, ignore_index=True)
        
```
